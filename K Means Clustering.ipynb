```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#To make fake data I use make_blobs Algorithm

from sklearn.datasets import make_blobs
```

```python
blobs = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.0,shuffle=True,random_state=101)
```

```python
blobs[0].shape
```

```python
plt.scatter(blobs[0][:,0],blobs[0][:,1])
```

```python
#Actual Lables
blobs[1]
```

```python
plt.scatter(blobs[0][:,0],blobs[0][:,1], c=blobs[1],cmap='rainbow')
```

```python
#Here import the Unsupervised Learning algorithm Kmeans

from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(blobs[0])
```

```python
kmeans.cluster_centers_
```

```python
#Predicted Labels
kmeans.labels_
```

```python
fig, (ax1,ax2)=plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(blobs[0][:,0], blobs[0][:,1], c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Orignal')
ax2.scatter(blobs[0][:,0], blobs[0][:,1], c=blobs[1],cmap='rainbow')
```
